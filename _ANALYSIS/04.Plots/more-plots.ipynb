```python
dfpop = pop.groupby(pop["Region"]).sum()
dfpop
```

```python
Poop = dfpop.transpose()
Poop
```

```python
poop = Poop.reset_index()
poop
```

```python
Poopi = poop.set_index([[2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]])
Poopi
```

```python
dfmed = med.groupby(med["Region"]).mean()
dfmed
```

```python
MED = dfmed.transpose()
MED
```

```python
Meed = MED.reset_index()
Meed
```

```python
meed = Meed.set_index([[2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]])
meed
```

```python
meed.plot(kind='line', marker = 'o')
plt.title("Average Median Income Per Region")
plt.ylabel("Income $")
plt.xlabel("Year")
plt.savefig("../Images/Avg_Median.png")

plt.show()
```

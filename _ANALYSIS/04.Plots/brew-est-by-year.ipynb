```python
# Dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
# Load in CSV

brew_df = pd.read_csv("../CSVs/census-brewery-region-data.csv")
brew_df.head()
```

```python
# Removing Duplicates

brew_df.drop_duplicates(subset = "Brewery Name", keep='first', inplace = True)
brew_df.head()
```

```python
brew_df = brew_df[(brew_df["Year Established"] >= 2010) & (brew_df["Year Established"] <= 2017)]
brew_df.head()
```

```python
# Create Groups for Year Established

est_group = brew_df.groupby('Year Established')

# Count how many times each Year Established appears
count_est = est_group['Year Established'].count()

count_est
```

```python
# Create a bar chart based off the group series

est_chart = count_est.plot(kind='bar', colormap="copper_r")

# Set xlabel & ylabel & title

plt.title("Breweries Established by Year")
plt.xlabel("Year Established")
plt.ylabel("Number of Breweries")

plt.show()
plt.tight_layout()
```

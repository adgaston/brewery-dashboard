```python
# Dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
# Load in CSV

brew_df = pd.read_csv("../CSVs/census-brewery-region-data.csv")
brew_df.head()
```

```python
# Removing Duplicates

brew_df.drop_duplicates(subset = "Brewery Name", keep='first', inplace = True)
brew_df.head()
```

```python
brew_df = brew_df[(brew_df["Year Established"] >= 2010) & (brew_df["Year Established"] <= 2017)]
brew_df.head()
```

```python
pd.crosstab(brew_df['Region'],brew_df['Still in Business?']).plot(kind="bar", colormap="copper")

plt.title("Breweries Still in Business by Region")
plt.xlabel("Region")
plt.ylabel("Number of Breweries")

plt.show()
plt.tight_layout()
```

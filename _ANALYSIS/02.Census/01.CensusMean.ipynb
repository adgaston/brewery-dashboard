```python
import pandas as pd
import matplotlib as mp
import numpy as np
```

```python
cen = "../CSVs/census_df_clean.csv"
brew = "../CSVs/brewerydb_final.csv"
```

```python
Cen = pd.read_csv(cen)
Brew = pd.read_csv(brew)
```

```python
cen_ = Cen.groupby(["State"]).sum()
cen_.head()
```

```python
mean = cen_.mean()
mean
```

```python
dfmean = mean[["2010 POP", "2011 POP", "2012 POP", "2013 POP", "2014 POP", "2015 POP", "2016 POP", "2017 POP"]]
dfmean.head()
```

```python
dfPOP = mean[["Median 2010", "Median 2011", "Median 2012", "Median 2013", "Median 2014", "Median 2015", "Median 2016", "Median 2017"]]
dfPOP.head()
```
